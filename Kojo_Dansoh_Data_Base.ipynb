# Let's take a look at one county (Baltimore City) in the data set.

```
 baltimore = df[df['Jurisdiction'] == 'Baltimore City']
baltimore
```

I ran this data by picking out one city and this table explains all the ED visits in Baltimore City, as you can see each City above have 50 rows for the data set and in the 50 rows start from 2008,and are divided into 5 rows for each year Until 2017.
The rows are broken down by year and by race.
The Value represents the number of ED admission


# Group By

```
baltimore.groupby('Year')['Value'].describe()
```

# Its seem like the mean of the ED admission rate for diabetes did not decrease through out years in the county of balitmore 

This code I ran is grouping the how many ED admission has happen throughout the decade and it is also divided by year. The mean shows average of total ED due to diabetes by year. According to this data it shows that the total average amount of ED visits has been increasing from 2008 until 2017

```
baltimore.groupby('Race/ ethnicity')['Value'].describe()
```

This code shows the mean average ED visit amongst race over the last decade. 

```
baltimore.groupby(['Race/ ethnicity','Year'])['Value'].describe()
```

The table above represents the average number of ED visit by race through the out years. It is interseting because it this shows that average of ED visit from blacks Non-Hispanic is greater than white Non-Hispanic. In addtion there was not data collected on Hispanics.
